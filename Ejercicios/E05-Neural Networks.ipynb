### Caperucita Roja usando Keras




````python
#Cargando librerias a utilizar 
from keras.models import Sequential
from keras.layers.core import Dense
````

````python
#Se genera la matriz 
X = np.array([[1,1,0,0],[0,1,1,0],[0,0,0,1]], "float32")
Y = np.array([[1,0,0,0],[0,0,1,1],[0,1,1,0]], "float32")
````

````python
#Se entrena el modelo con Keras
Modk = Sequential()
Modk.add(Dense(16, input_dim=4, activation='relu'))
Modk.add(Dense(4, activation='sigmoid'))
````

````python
Modk.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])
````

````python
Modk.fit(X, Y, epochs=1000)
````

````python
scores = Modk.evaluate(X, Y)
 
print((Modk.metrics_names[1], scores[1]*100))
print (Modk.predict(X).round())
````

##  4.2 Boston House Price Prediction

In the next questions we are going to work using the dataset *Boston*. This dataset measures the influence of socioeconomical factors on the price of several estates of the city of Boston. This dataset has 506 instances, each one characterized by 13 features:

* CRIM - per capita crime rate by town
* ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
* INDUS - proportion of non-retail business acres per town.
* CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
* NOX - nitric oxides concentration (parts per 10 million)
* RM - average number of rooms per dwelling
* AGE - proportion of owner-occupied units built prior to 1940
* DIS - weighted distances to five Boston employment centres
* RAD - index of accessibility to radial highways
* TAX - full-value property-tax rate per 10,000 USD
* PTRATIO - pupil-teacher ratio by town
* B - $1000(Bk - 0.63)^2$ where $Bk$ is the proportion of blacks by town
* LSTAT - % lower status of the population

Output variable:
* MEDV - Median value of owner-occupied homes in 1000's USD

**Note:** In this exercise we are going to predict the price of each estate, which is represented in the `MEDV` variable. It is important to remember that we are always aiming to predict `MEDV`, no matter which explanatory variables we are using. ## 4.4 Modeling linear and non linear relationships

* Generate two new subsets filtering these characteristics:
     * $D_1$:  $X = \textit{'RM'}$, $y = \textit{'MEDV'}$
     * $D_2$:  $X = \textit{'LSTAT'}$, $y = \textit{'MEDV'}$
* For each subset, generate a training partition and a test partition using a ratio of $ 70 \% - 30 \% $
* Train a linear regression model on both subsets of data:

#### 1. Generar dos nuevos subconjuntos D1 y D2 

````python
#Generar dos nuevos subconjuntos D1 y D2 
D1= pd.DataFrame(Boston.loc[:,["RM","MEDV"]])
D2= pd.DataFrame(Boston.loc[:,["LSTAT","MEDV"]])
````

````python
#Se muestra como quedo el subconjunto D1 y D2
D1.head()
````

````python
D2.head()
````

#### 2. Generar una partición de entrenamiento y una partición de prueba usando una proporción de 70% -30% para cada subconjunto

````python
from sklearn.model_selection import train_test_split
D1_X_train, D1_X_test, D1_y_train, D1_y_test = train_test_split(D1["RM"], D1["MEDV"], random_state=950, test_size=0.3)
D2_X_train, D2_X_test, D2_y_train, D2_y_test = train_test_split(D2["LSTAT"], D2["MEDV"], random_state=950, test_size=0.3)
````

#### 3- Entrenar un modelo de regresión en cada uno de los subconjuntos 

````python
#Cargando las librerias a utilizar 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
````

````python
D1_y_test=np.array(D1_y_test)
D2_y_test=np.array(D2_y_test)
````

````python
#Definir el objeto de cada subconjunto 
D1lr = LinearRegression()
D2lr = LinearRegression()
````

````python
#Entrenamiento del modelo 
D1lr.fit(D1_X_train.values.reshape(-1, 1),D1_y_train.values.reshape(-1, 1))
D2lr.fit(D2_X_train.values.reshape(-1, 1),D2_y_train.values.reshape(-1, 1))
````

````python
# Predecir en Conjunto de Test
pred_D1_y=D1lr.predict(D1_X_test.values.reshape(-1, 1))
pred_D2_y=D2lr.predict(D2_X_test.values.reshape(-1, 1))
````

````python
#Imprimir el MSE 
from sklearn.metrics import mean_squared_error
print("MSE_D1: %.2f"% mean_squared_error(D1_y_test, pred_D1_y))
print("MSE_D2: %.2f"% mean_squared_error(D2_y_test, pred_D2_y))
````

````python
#Imprima los valores de w y w0 de la ecuación de regresión

print('Intercepto D1: ',D1lr.intercept_,)
print('Coeficiente D1: ',D1lr.coef_,)

print('Intercepto D2: ',D2lr.intercept_,)
print('Coeficiente D2: ',D2lr.coef_,)
````

````python
#gráfico donde visualice la línea obtenida por el modelo de regresión 
#junto con los datos de entrenamiento y los datos de prueba
import matplotlib.pyplot as plt
D1.plot(x='RM',y='MEDV',kind ='scatter')
plt.plot(D1['RM'],D1lr.intercept_+D1lr.coef_[0]*D1['RM'], color ='green')
````

````python
D2.plot(x='LSTAT',y='MEDV',kind ='scatter')
plt.plot(D2['LSTAT'],D2lr.intercept_+D2lr.coef_[0]*D2['LSTAT'], color ='green')
````

#### 3. Como funciona el modelo en D1 y D2.  Por qué?

El modelo en D1 tiene mayor MSE en la muestra de test que el modelo D2.
Con D1 se aprecian varios puntos que no siguen la tendencia lineal que se esta ajustando, y no es bueno prediciendo en los extremos.
En D2 la tendencia no es necesariamente lineal, más bien es curva a medida que la proporción de población de bajos ingresos aumenta por encima de 25%, el precio de la vivienda no continúa bajando y se estabiliza.

Ambos modelos tienen un bajo desempeño, porque no consideran informacion adicional y aunque puedea haber una relacion monotona con la variable respuesta, esta no es necesariamente lineal.
